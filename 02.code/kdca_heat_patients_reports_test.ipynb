```python
import os

import pandas as pd
import numpy as np
import scipy as spp

import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns
```

```python
# 나눔 폰트 경로 설정
# dir = '/data1/E382/hwchoi/00.data/00.preprocessing'
dir='C:/Windows/Fonts'
font_path = os.path.join(dir, 'NANUMGOTHIC.TTF')
# font_path = os.path.join(dir, 'DejaVu Sans')

# 폰트 속성 설정
fontprop = fm.FontProperties(fname=font_path, size=10)


# 폰트 매니저에 폰트 추가
fm.fontManager.addfont(font_path)

# Matplotlib 폰트 캐시 파일 삭제 (필요시 사용)
cache_dir = os.path.expanduser('~/.cache/matplotlib')
font_cache_path = os.path.join(cache_dir, 'fontlist-v330.json')
if os.path.exists(font_cache_path):
    os.remove(font_cache_path)

# 폰트를 matplotlib에 설정
plt.rc('font', family=fontprop.get_name())
```

```python
df=pd.read_csv('D:/Work/002.code/jupyterNotebook/data/SUB_ANAL/patients_analy_lst_0822.csv')
df.head(10)
```

```python
df[df['나이']>=100]
```

```python
# 성별에 따른 온열질환자 분포 시각화
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='성별', palette='Set2')
plt.title('성별에 따른 온열질환자 분포')
plt.xlabel('성별')
plt.ylabel('환자 수')
plt.show()
```

```python
# 나이에 따른 분포 시각화
plt.figure(figsize=(8, 6))
# sns.histplot(data=df, x='나이', bins=10, kde=True, color='blue')
sns.histplot(data=df, x='나이', bins=10)
plt.title('나이에 따른 온열질환자 분포')
plt.xlabel('나이')
plt.ylabel('환자 수')
plt.show()
```

```python
# 발생년월에 따른 분포 시각화
plt.figure(figsize=(15, 6))
sns.countplot(data=df, x='발생년월', palette='Set1', order=sorted(df['발생년월'].unique()))
plt.title('발생년월에 따른 온열질환자 분포')
plt.xlabel('발생년월')
plt.ylabel('환자 수')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.histplot(data=df, x='나이', hue='성별', bins=10, multiple='stack', palette='Set2')
plt.title('온도에 따른 온열질환자 분포')
plt.xlabel('연령대')
plt.ylabel('환자 수')
plt.show()
```
